```python
import sys
sys.path.insert(0, '../')
from numerical_representation_strategies.physicochemical_properties import physicochemical_encoder
from numerical_representation_strategies.fft_encoding import fft_encoding
from numerical_representation_strategies.constant_values import constant_values

import pandas as pd
```

```python
df_data = pd.read_csv("../../dataset_examples/demo_df_sequences.csv")
df_data
```

```python
property_to_use = 'ANDN920101'
encoders = pd.read_csv("../../inputs_encoder/aaindex_encoders.csv")
encoders.index = encoders['residue']

column_with_seq = 'sequence'
column_with_response = 'peak'
```

```python
print("Instance object")
encoding_instance = physicochemical_encoder(df_data,
                 property_to_use,
                 encoders,
                 constant_values(),
                 column_with_seq,
                 column_with_response)
```

```python
df_data_encoding = encoding_instance.encoding_dataset()
df_data_encoding
```

```python
response_column = df_data_encoding['peak']
df_data_to_train = df_data_encoding.drop(columns=['peak'])
```

```python
from training_classic_ml.regx_ml_models import regression_models

regx_models = regression_models(df_data_to_train, response_column)
X_train, X_test, y_train, y_test = regx_models.split_to_train(test_size=0.2)

regx_models.apply_model(X_train=X_train, y_train=y_train, option=1)

performances = regx_models.get_performances(X_test=X_test, y_test=y_test)

performances
```
